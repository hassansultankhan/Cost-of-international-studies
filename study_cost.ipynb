```python
top_continents = df['Continent'].value_counts()
top_continents

plt.figure(figsize=(10, 6))
sns.barplot(x=top_continents.index, y=top_continents.values, palette='viridis')
plt.title('Top Continents by Number of Application')
plt.xlabel('Continent')
plt.ylabel('Number of Applications')
plt.show()

```
